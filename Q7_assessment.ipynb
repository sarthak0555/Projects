```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import pandas as pd
```

```python
# Load the modelling_data
modelling_data = pd.read_excel('/Users/sarthakchawla/Downloads/modelling_data.xlsx')
```

```python
modelling_data
```

```python
# Separate the dependent variable (y) and the independent variables (x1-x100)
y = modelling_data['y']
X = modelling_data.drop(columns=['y'])
```

```python
# Initialize variables to track the best model
best_r2 = -1
best_i = None
```

```python
# Iterate over i in [3, 4, ..., 100]
for i in range(3, 101):
    # Select the independent variables x1, x2, and xi
    X_selected = X[['x1', 'x2', f'x{i}']]
    
    # Create and fit the linear regression model
    model = LinearRegression()
    model.fit(X_selected, y)
    
    # Predict the dependent variable
    y_pred = model.predict(X_selected)
    
    # Calculate the R-Squared value
    r2 = r2_score(y, y_pred)
    
    # Update the best model if the current model has a higher R-Squared value
    if r2 > best_r2:
        best_r2 = r2
        best_i = i

best_i, best_r2
```
