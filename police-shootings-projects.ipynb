## 2: Plot 12 most dangerous cities

```python
df.head()
```

```python
df.isnull().sum()
```

```python
most_dangerous_cities=df.city.value_counts()[:12]
a=most_dangerous_cities.sort_values()
a
```

```python
sns.barplot(data=a, x=a.index,y=a.values)
```

```python
most_dangerous_cities.plot(kind="bar")
```

```python
x=most_dangerous_cities.index
x
```

```python
y=most_dangerous_cities.values
y
```

```python
plt.figure(figsize=(12,6),dpi=100)
sns.barplot( x,y)
plt.xticks(rotation=60)
plt.xlabel("cities")
plt.title("most dangerous cities",color="b")
plt.ylabel("count")

for index , value in enumerate(y):
    plt.text(index, value+0.1, str(value), ha = 'center')
```

```python
most_dangerous_cities.index
```

```python
# with seaborn countplot
plt.figure(figsize=(12,6))
sns.countplot(data=df,x="city",order=most_dangerous_cities.index)
plt.xticks(rotation=60)
plt.show();
```
